```python
import nltk
import json
from functools import reduce
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

with open('ABTParsedParagraphs.json','r', encoding='utf-8') as f:
    paragraphs_with_min_words=json.load(f)
```

```python
%%time
sentences = reduce(lambda x,y: x+sent_tokenize(y), [[]]+paragraphs_with_min_words)
```

```python
import json
with open('ABTParsedSentences.json','w', encoding='utf-8') as f:
    json.dump(paragraphs_with_min_words, f, ensure_ascii=False)
```
